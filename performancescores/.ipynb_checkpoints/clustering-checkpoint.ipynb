```python
from scipy.io import mmread,mminfo
from scipy.sparse import csr_matrix
import numpy as np
import networkit as nk
import simexpal
import yaml
import pandas

def readEmbeddings(filename, nodes, dim):
    embfile = open(filename, "r")
    firstline = embfile.readline()
    N = int(firstline.strip().split()[0])
    X = [[0]*dim for i in range(nodes)]
    setrange = nodes
    count = 0
    for line in embfile.readlines():
        tokens = line.strip().split()
        nodeid = int(tokens[0])-1
        x = []
        for j in range(1, len(tokens)):
            t = float(tokens[j])
            x.append(t)
        if nodeid >= nodes:
            continue
        X[nodeid] = x
        count += 1
        if count > setrange:
            break
    embfile.close()
    return np.array(X)

def readFile(filename):
    print("Read graph: ", filename)
    G = nk.readGraph(filename, nk.Format.NetworkitBinary)
    return G

def calculateQuality(G, embeddings):
    partition = nk.community.PLM(G, refine=True, maxIter=100).run().getPartition()
    print("Baseline (PLM) computed.")
    clusterIDs = partition.getSubsetIds()

    totalDist = np.float64(0.0)
    for id in clusterIDs:
        member = list(partition.getMembers(id))
        for i in range(0, len(member)):
            for j in range(i+1, len(member)):
                totalDist += np.linalg.norm(embeddings[member[i]]-embeddings[member[j]])
    print("Total dist: ", totalDist)
    return totalDist

def calculateQuality2(G, embeddings, partition):
    clusterIDs = partition.getSubsetIds()

    totalDist = np.float64(0.0)
    for id in clusterIDs:
        member = partition.getMembers(id)
        for m in member:
            for neighbor in G.iterNeighbors(m):
                if partition.subsetOf(neighbor) == id:
                    totalDist += np.linalg.norm(embeddings[neighbor]-embeddings[m])
    print("Total dist: ", totalDist)
    return totalDist

def parse(run, G, partition1, partition2, filter = 'all'):
    # Get all used parameter
    var_dict = dict()
    for var in successful_run.experiment.variation:
        var_dict[var.axis] = var.name[1:]
    
    if filter != 'all':
        if var_dict['batch'] not in filter:
            return

    # Read output file
    X = readEmbeddings(run.output_file_path("embd"), G.numberOfNodes(), 128)
    
    # Calculate quality
    q1 = calculateQuality2(G, X, partition1)
    q2 = calculateQuality2(G, X, partition2)
    
    return {
        'experiment': run.experiment.name,
        'instance': run.instance.shortname,
        'option': var_dict['option'],
        'batch': var_dict['batch'],
        'iter': var_dict['iter'],
        'dist_plm': q1,
        'dist_ple': q2
    }

def parse2(run, G, partition1, partition2, filter = 'all'):
    # Get all used parameter
    var_dict = dict()
    for var in successful_run.experiment.variation:
        var_dict[var.axis] = var.name[1:]
    
    if filter != 'all':
        if var_dict['batch'] not in filter:
            return

    # Read output file
    X = readEmbeddings(run.output_file_path("embd"), G.numberOfNodes(), 128)
    
    # Calculate quality
    q1 = calculateQuality2(G, X, partition1)
    q2 = calculateQuality2(G, X, partition2)
    
    return {
        'experiment': run.experiment.name,
        'instance': run.instance.shortname,
        'option': var_dict['wl'],
        'batch': var_dict['batch'],
        'dist_plm': q1,
        'dist_ple': q2
    }

```

```python
cfg = simexpal.config_for_dir("../")
results = []
G = readFile('/work/brandtfa/2023/force2vec/fork/force2vec/datasets/input/com-youtube.nkb')
partition1 = nk.community.PLM(G, refine=True, maxIter=100).run().getPartition()
partition2 = nk.community.ParallelLeiden(G,iterations=100).run().getPartition()
```

```python
for successful_run in cfg.collect_successful_results():
    results.append(parse(successful_run, G, partition1, partition2))

df = pandas.DataFrame(results)
df.to_csv('data.csv')
print(df)
```

```python
import matplotlib.pyplot as plt

G = readFile('/work/brandtfa/2023/force2vec/fork/force2vec/datasets/input/com-youtube.nkb')
res_df = pandas.read_csv('data.csv')

# normalize distances
n = G.numberOfNodes()
minDist = min(min(res_df['dist_plm']), min(res_df['dist_ple']))
res_df['dist_plm'] = res_df['dist_plm'] / minDist
res_df['dist_ple'] = res_df['dist_ple'] / minDist

# create filters: algos, iterations
sig = res_df['option']=='VX-sigmoid-rw'
tdist = res_df['option']=='VX-tdist-ns'
res_df_tdist = res_df[tdist]
res_df_sig = res_df[sig]
tdist500 = res_df_tdist['iter']==500
tdist1000 = res_df_tdist['iter']==1000
tdist2000 = res_df_tdist['iter']==2000
tdist5000 = res_df_tdist['iter']==5000
sig500 = res_df_sig['iter']==500
sig1000 = res_df_sig['iter']==1000
sig2000 = res_df_sig['iter']==2000
sig5000 = res_df_sig['iter']==5000

```
